## Matplotlib Example Workflow

```python
import matplotlib.pyplot as plt

# Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup Plot
fig, ax = plt.subplots(figsize=(5, 3)) # (width, height)

# Plot Data
ax.plot(x, y)

# Customize Plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save & Show
fig.savefig("sample-plot.png")
```

### Bar plot

```python
# Plot from dictionary

butter_prices = {"Almond butter": 10,
                "Peanut butter": 8,
                "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values()) # (x, y)

ax.set(title = "Nut Butter Store",
      ylabel = "Prices($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))

# for horizontal bar plots, parameters must be converted to Python lists

ax.set(title = "Nut Butter Store",
      xlabel = "Prices($)");
```

### Histogram

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Combined Subplots

```python
# Option 1 (Recommended)

x = np.linspace(0, 20, 100)

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

ax0.plot(x, np.sin(x))
ax1.scatter(np.random.random(10), np.random.random(10))
ax2.bar(butter_prices.keys(), butter_prices.values())
ax3.hist(np.random.randn(100));
```

```python
# Option 2

fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))

ax[0,0].plot(x, np.sin(x))
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(butter_prices.keys(), butter_prices.values())
ax[1,1].hist(np.random.randn(100));
```

### Pyplot vs OO method
* For quick plotting, use pyplot method
* For advanced plotting, use OO method

```python
hd.head()
```

```python
over50 = hd[hd["age"] > 50]
over50.head()
```

```python
len(over50)
```

#### Pyplot Method

```python
over50.plot.scatter(figsize = (10, 6),
                    x = "age",
                    y = "chol",
                    c = "target");
```

#### OO Method mixed with pyplot

```python
over50.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over50.plot.scatter(x = 'age',
                    y = 'chol',
                    c = 'target',
                    ax = ax);

# ax.set_xlim([45, 100])
```

#### OO Method from scratch 

```python
over50.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

sc = ax.scatter(x = over50['age'],
                y = over50['chol'],
                c = over50['target'])

# Customization 
ax.set(title = "Heart disease and Cholesterol level",
        xlabel = "Age",
        ylabel = "Cholesterol")

ax.legend(*sc.legend_elements(), title = "Target")

ax.axhline(over50['chol'].mean(),
          linestyle = "--");
```

```python
over50.head()
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)     # both axes will share the x-axis (age)

# chol vs age
sc = ax0.scatter(x = over50["age"],
                y = over50["chol"],
                c = over50["target"])
ax0.set(title = "Heart disease & Cholesterol Level",
       ylabel = "Cholesterol")
ax0.legend(*sc.legend_elements(), title = "Target")
ax0.axhline(y = over50["chol"].mean(),
           linestyle = "--")

# thalach vs age
sc = ax1.scatter(x = over50["age"],
                y = over50["thalach"],
                c = over50["target"])
ax1.set(title = "Heart Disease & Max Heart Rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate")
ax1.legend(*sc.legend_elements(), title = "Target")
ax1.axhline(y = over50["thalach"].mean(),
           linestyle = "--");

# Adding a super title
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
plt.style.use("seaborn-whitegrid")

# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)     # both axes will share the x-axis (age)

# chol vs age
sc = ax0.scatter(x = over50["age"],
                y = over50["chol"],
                c = over50["target"],
                cmap = "winter") # this changes the color scheme
ax0.set(title = "Heart disease & Cholesterol Level",
       ylabel = "Cholesterol")
ax0.legend(*sc.legend_elements(), title = "Target")
ax0.axhline(y = over50["chol"].mean(),
           linestyle = "--",
           c = "red")
ax0.set_xlim([50, 80])

# thalach vs age
sc = ax1.scatter(x = over50["age"],
                y = over50["thalach"],
                c = over50["target"],
                cmap = "autumn")
ax1.set(title = "Heart Disease & Max Heart Rate",
       xlabel = "Age",
       ylabel = "Max Heart Rate")
ax1.legend(*sc.legend_elements(), title = "Target")
ax1.axhline(y = over50["thalach"].mean(),
           linestyle = "--",
           c = "blue");
ax1.set_ylim([60, 200])

# Adding a super title
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
