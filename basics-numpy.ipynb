#### Standard Deviation & Variance

```python
# Standard Deviation = a measure of how spread out a group of numbers is from the mean
#                    = sqrt(variance)

np.std(a2)
```

```python
# Variance = measure of the avg degree to which each number is different to the mean
# Higher variance = wider range of numbers
# Lower variance = lower range of numbers

np.var(a2)
```

### Dot Product Practical Example

```python
np.random.seed(0)

# Number of jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# Weekly sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,
                           index=["Mon", "Tue", "Wed", "Thu", "Fri"],
                           columns=["Almond", "Peanut", "Cashew"])
weekly_sales
```

```python
# Prices array
prices = np.array([10,8,12])
prices
```

```python
# Prices in DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),
                            index=["Price"],
                            columns=["Almond", "Peanut", "Cashew"])
butter_prices
```

The array `prices` has to be reshaped. Guess why?

```python
weekly_sales, butter_prices
```

```python
total_sales = prices.dot(sales_amounts.T) # sales_amount is transposed here
total_sales
```

```python
butter_prices
```

```python
weekly_sales
```

```python
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```

```python
weekly_sales.to_csv("weekly-sales.csv")
```
